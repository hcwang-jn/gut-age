```r
library(ggplot2)
```

```r
dat = read.csv('Data/1.sp_shannon.csv')

options(repr.plot.width=5, repr.plot.height=2)
ggplot(dat, aes(age, shannon, color=age))+
    geom_point(alpha=0.1, size=0.5)+
    theme_bw()+
    geom_smooth(color='#B9BEBA', alpha=0.5)+
    xlab('Age')+
    ylab('Shannon')+
    scale_color_gradient(low = '#2295C7', high = '#2F56B5')+
    labs(color='Age')
```

```r
dat = read.csv('Data/3.sp_bray_pcoa_result.csv')

options(repr.plot.width=4, repr.plot.height=4)
ggplot(dat, aes(Axis.1, Axis.2, color=age))+
    geom_point(alpha=0.2, size=.5)+
    theme_bw()+
    scale_color_gradient2(low = '#2295C7', mid = '#94C280', high = '#F1C359', midpoint = 60)+
    labs(color='Age')+
    xlab('PCo 1')+
    ylab('PCo 2')+
    theme(legend.position = c(.12, .77),
          legend.box.background = element_rect(color="black", size=1))+
    guides(color = guide_colorbar(barwidth = .8, barheight = 5))
```
