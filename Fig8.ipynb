```r
library(ggplot2)
library(pheatmap)
```

```r
spr = read.csv('Data/25.n30_cor_spr.csv', row.names = 1)
pvs = read.csv('Data/26.sig_marker.csv', row.names = 1)

grp = c(rep('scale', 2), rep('disease', 1), rep('plasma', 21))
grp = as.data.frame(grp)
rownames(grp) = colnames(spr)

clf = c(rep('Pos', 15), rep('Neg', 15))
clf = as.data.frame(clf)
rownames(clf) = rownames(spr)

pheatmap(t(spr), cluster_cols = F, cluster_rows = F,
         color = colorRampPalette(c('#36A1DF', 'white', '#E3282F'))(200),
         display_numbers = t(pvs), annotation_row=grp, annotation_col=clf)
```
