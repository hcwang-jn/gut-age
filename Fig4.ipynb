```r
library(ggplot2)
library(ggbeeswarm)
```

```r
dat = read.csv('Data/13.all_region_explain_beeswarm_top20_plot.csv')
rFv2 = (dat$Fv - min(dat$Fv)) / (max(dat$Fv) - min(dat$Fv))

options(repr.plot.width=5, repr.plot.height=3)
ggplot(dat, aes(ALE, reorder(tFid, ALE, mean), color=rFv2))+
    theme_bw()+
    geom_quasirandom(alpha=.3, size=.6, groupOnX = F)+
    scale_color_gradient2(low='#0C73B2', mid='#DE6A63', high = '#A90504', midpoint = 0.5)+
    ylab('Feature')+
    labs(color='Feature\nValues')+
    theme(axis.text.y = element_text(angle = 0, hjust = 1, vjust = .5, size = rel(.8)))+
    theme(legend.position = c(0.87,0.28))+
    theme(legend.key.size = unit(.4, 'cm'))+
    theme(legend.title = element_text(size=rel(.8)))
```
