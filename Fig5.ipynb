```r
library(ggplot2)
library(ggsignif)
```

```r
data = read.csv('Data/14.disease_number_with_age_predage.csv')

options(repr.plot.width=4, repr.plot.height=2)
ggplot(data, aes(factor(n_disease), Age, fill=n_disease))+
    geom_boxplot(outlier.color = NA)+
    theme_bw()+
    guides(fill='none')+
    xlab('Number of diseases')+
    ylab('Chronological Age')+
    scale_fill_gradient2(low = '#459BB7', mid = '#F2F3F4', high = '#D8451D', midpoint = 2.5)
```

```r
options(repr.plot.width=4, repr.plot.height=2)
ggplot(data, aes(factor(n_disease), Pred, fill=n_disease))+
    geom_boxplot(outlier.color = NA)+
    theme_bw()+
    guides(fill='none')+
    xlab('Number of diseases')+
    ylab('Predicted Age')+
    scale_fill_gradient2(low = '#459BB7', mid = '#F2F3F4', high = '#D8451D', midpoint = 2.5)
```

```r
data = read.csv('Data/15.pred_residual_with_different_health_status.csv')

options(repr.plot.width=2, repr.plot.height=3)
ggplot(data, aes(reorder(disease, residual, median), residual, fill=disease))+
    geom_boxplot(outlier.color = NA)+
    theme_bw()+
    geom_hline(yintercept=0, color='grey', linetype='longdash', alpha=.5)+
    geom_signif(map_signif_level = T,
                step_increase = 0.1,
                comparisons = list(
        c('IBD', 'control'), c('T2D', 'control'), c('control', 'cirrhosis'),
        c('control', 'adenoma'), c('control', 'CRC')),
    )+
    guides(fill='none')+
    xlab('Disease Status')+
    ylab('Predict Residual')+
    scale_fill_manual(values = c('#FFD0A1', '#7DE3D1', '#3D9FD3', 
                                 '#AFADDE', '#C56E90', '#80BFA0'))+
    theme(axis.text.x = element_text(angle=90, hjust = 1, vjust = .5))
```

```r
data = read.csv('Data/16.ind_disease_csp_pred.csv')

options(repr.plot.width=4, repr.plot.height=4)
ggplot(data, aes(FPR, TPR, group=Group))+
    theme_bw()+
    geom_ribbon(aes(ymin=TPRMIN, ymax=TPRMAX, fill=Group), alpha=.1)+
    geom_path(aes(color=Group))+
    geom_abline(intercept=0,slope=1, color='black', alpha=0.8, linetype='longdash')+
    scale_fill_manual(values = c('#FFD0A1', '#7DE3D1', '#3D9FD3', 
                                 '#AFADDE', '#C56E90', '#80BFA0'),
                      labels = c('Adenoma (Mean AUC 0.56)',
                                 'Cirrhosis (Mean AUC 0.93)',
                                 'CRC (Mean AUC 0.81)',
                                 'IBD (Mean AUC 0.97)',
                                 'T2D (Mean AUC 0.69)'))+
    scale_color_manual(values = c('#FFD0A1', '#7DE3D1', '#3D9FD3', 
                                 '#AFADDE', '#C56E90', '#80BFA0'),
                      labels = c('Adenoma (Mean AUC 0.56)',
                                 'Cirrhosis (Mean AUC 0.93)',
                                 'CRC (Mean AUC 0.81)',
                                 'IBD (Mean AUC 0.97)',
                                 'T2D (Mean AUC 0.69)'))+
    theme(legend.position = c(.68, .2))+
    theme(legend.background = element_rect(fill=NA, color='black'))+
    theme(legend.key.size = unit(12, 'pt'))+
    xlab('True Positive Rate')+
    ylab('False Positive Rate')
```

```r
data = read.csv('Data/18.ind_pathway_marker_heatmap_plot_lite.csv')

options(repr.plot.width=8, repr.plot.height=4)
ggplot(data, aes(Marker, factor(Group, levels = c('CRC', 'Adenoma', 'Cirrhosis', 'IBD', 'T2D', 'gAge')), 
                 fill=factor(Value)))+
    geom_tile()+
    theme_bw()+
    theme(axis.text.x = element_text(angle=60, hjust = 1, vjust = 1))+
    scale_fill_manual(values = c('#F6F6F7', '#3CB7CB'),
                      labels = c('Non-marker', 'Marker'))+
    theme(legend.position = 'bottom')+
    labs(fill='Group')+
    ylab('Status')
```
