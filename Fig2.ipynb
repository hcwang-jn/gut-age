```r
dat = read.csv('Data/5.species_species_stats_barplot.csv', row.names=1)

options(repr.plot.width=9, repr.plot.height=4)
ggplot(dat, aes(Age, Relab, fill=Species))+
    geom_bar(stat="identity", alpha=1, width = 1, space=0)+
    scale_x_continuous(limits=c(18,92), breaks=seq(20, 90, 10))+
    theme_bw()+
    scale_fill_manual(values = c('#FBA120', '#F2D670', '#FEECD4', '#C4BDC9', '#D8D8D8', '#44CFCC', '#A64435', '#7E9A8A', '#C0BF92', '#5BB9BD', '#1D8FAD', '#235D92', '#E3C2AA', '#C0E7C5',
                                 '#83BEC0'))+
    theme(legend.position = "right", legend.key.size = unit(15, 'pt'))+
    ylab('Relative Abundance')
```

```r
dat = read.csv('Data/6.pathway_pwy_spearman_barplot.csv')
dat$spr = 'Spearman\'s Rho'

options(repr.plot.width=9, repr.plot.height=1.9)
ggplot(dat, aes(X, spr, fill=Spearman.s.Rho))+
    geom_tile()+
    theme_bw()+
    theme(axis.text.x = element_text(angle=30, hjust=1, vjust=1))+
    scale_fill_gradient(low = '#1D8FAD', high = '#DDE7EE')+
    xlab('Pathway')+
    ylab('')+
    labs(fill='')+
    guides(fill=guide_colorbar(direction = 'horizontal'))
```
