```r
library(ggplot2)
```

```r
dat2 = list(rep(c('ADL', 'MoCA', 'MMAS-8'),2))
dat2 = as.data.frame(dat2)
colnames(dat2) = 'index'

dat2$spr = c(0.20192992244271438, -0.06760702021414433, 0.2663937963060426, 0.01664583020081491,
             -0.24369072149919693, 0.11894287347566089)
dat2$type = c(rep(c('gAge'), 3), rep(c('Chronological Age'), 3))

options(repr.plot.width=3.3, repr.plot.height=1.8)
ggplot(dat2, aes(index, type, fill=spr))+
    geom_tile()+
    theme_bw()+
    scale_fill_gradient2(low = '#EC1B22', mid = '#CAE7E8', high = '#209BBA', midpoint = 0)+
    guides(fill='none')+
    geom_text(aes(label=round(spr,3)),size=3)+
    ylab('')+
    xlab('Score of aging')
```

```r
dat2 = list(rep(c('ADL', 'MoCA', 'MMAS-8'),2))
dat2 = as.data.frame(dat2)
colnames(dat2) = 'index'

dat2$r2 = c(0.03596, 0.02062, 0.0332, 0.05225, 0.02161, 0.03511)
dat2$type = c(rep(c('species'), 3), rep(c('pathway'), 3))

options(repr.plot.width=3.7, repr.plot.height=1.8)
ggplot(dat2, aes(reorder(index, -r2), r2, fill=type))+
    geom_bar(stat = 'identity', position = 'dodge')+
    theme_bw()+
    scale_fill_manual(values = c('#1C96BA', '#3CBCC7'))+
    ylab('Adonis R2')+
    xlab('Score of aging')
```

```r
dat = read.csv('Data/22.frail_cohort_pcoa_ggplot.csv')

options(repr.plot.width=2.5, repr.plot.height=2.5)
ggplot(dat, aes(Axis.1, Axis.2, color=grp))+
    geom_point()+
    stat_ellipse()+
    theme_bw()+
    guides(color='none')+
    xlab('PCo 1')+
    ylab('PCo 2')+
    scale_color_manual(values = c('#FD673A','#1C96BA'))
```

```r
dat = read.csv('Data/23.frail_cohort_overlap.csv')
dat$Scale = factor(dat$Scale, levels=rev(c('gAge', 'ADL', 'MoCA')))

options(repr.plot.width=8, repr.plot.height=1.3)
ggplot(dat, aes(new, Scale, fill=Scale))+
    geom_tile()+
    theme_bw()+
    scale_fill_manual(values=c("#A8E0D0", "#ACCCE8", "#01A8CA"))+
    guides(fill='none')+
    xlab('Samples')+
    ylab('Score of aging')+
    theme(axis.text.x = element_blank(),
          axis.ticks.x = element_blank())
```
